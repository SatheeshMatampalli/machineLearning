```python
import pandas as pd
data=pd.read_csv('olympics.csv',index_col=0,skiprows=1)
```

```python
for col in data.columns:
    if col[:2]=='01':
        data.rename(columns={col:'Gold'+col[5:]}, inplace=True)
    if col[:2]=='02':
        data.rename(columns={col:'Silver'+col[5:]}, inplace=True)
    if col[:2]=='03':
        data.rename(columns={col:'Bronze'+col[5:]}, inplace=True)
    if col[:1]=='№':
        data.rename(columns={col:'#'+col[1:]}, inplace=True)

names_ids = data.index.str.split('\s\(') # split the index b
data.index = names_ids.str[0] # the [0] element is the country name (new index) 
data['ID'] = names_ids.str[1].str[:3]
```
