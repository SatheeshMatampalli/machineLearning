```python
# total no of values
n=len(x)
# finding x and y means
num=0
denom=0
for i in range(n):
    num+=(x[i]-mean_x)*(y[i]-mean_y)
    denom+=(x[i]-mean_x)**2
b1=num/denom
b2=mean_y-(b1*mean_x)
print(b1)
print(b2)
```

```python
# ploting values and regression line
max_x=np.max(x)+100
min_x=np.min(x)-100
# calculation line values x and y
xx=np.linspace(min_x,max_x,1000)
yy=b2 + (b1 * xx)
# ploting line
plt.plot(xx,yy,color='red',label='Regression line')
# ploting scatter points
plt.scatter(x,y,c='orange',label='Scatter plot')
plt.xlabel('Head size in cm3')
plt.ylabel('Brain weight in brms')
plt.legend()
plt.show()
```

```python
ss_tot=0
ss_res=0
for i in range(n):
   y_pred=b2+b1 * x[i]
   ss_tot += (y[i] - mean_y) ** 2
   ss_res += (y[i]-y_pred) ** 2
print(y_pred)
r2=1-(ss_res/ss_tot)
print("R2_Score")
print(r2)
```
